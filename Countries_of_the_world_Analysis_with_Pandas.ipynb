# Feature Engineering

### - Add a column for GDP per capita:

```python
df['GDP_per_Capita']=df['GDP ($ per capita)']/df['Population']
```

```python
df.head()
```

### - Create a categorical column for population density

```python
# Creating a new column 'Density_category' based on a condition
df['Density_category'] = 'High'  # Initialize the 'Density_category' column with 'High' as the default value

# Using DataFrame.loc[] to set values based on the condition
df.loc[df['Pop. Density (per sq. mi.)'] < 50, 'Density_category'] = 'Low'
df.loc[(df['Pop. Density (per sq. mi.)'] >=50) & (df['Pop. Density (per sq. mi.)'] <=200), 'Density_category'] = 'Medium'
```

```python
df
```

### - Extract the continent from the region

```python
df['Continent']=np.nan
```

```python
df['Continent']='Others'
df.loc[(df['Region'] == 'ASIA (EX. NEAR EAST)') | (df['Region'] == 'NEAR EAST'), 'Continent'] = 'Asia'
df.loc[(df['Region']=='NORTHERN AFRICA')|(df['Region']=='SUB-SAHARAN AFRICA'),'Continent']='Africa'
df.loc[(df['Region']=='WESTERN EUROPE')|(df['Region']=='EASTERN EUROPE'),'Continent']='Europe'
df.loc[df['Region']=='NORTHERN AMERICA','Continent']='North America'
df.loc[df['Region']=='LATIN AMER. & CARIB','Continent']='South America'
df.loc[df['Region']=='OCEANIA','Continent']='Oceania'
df['Continent'].value_counts()
```

```python
df
```

### - Climate_category from Climate

```python
df['Climate_category']='Unknown'
df.loc[df['Climate'] == 1, 'Climate_category'] = 'Tropical'
df.loc[df['Climate'] == 2, 'Climate_category'] = 'Arid'
df.loc[df['Climate'] == 3, 'Climate_category'] = 'Temperate'
df.loc[df['Climate'] == 4, 'Climate_category'] = 'Cold'
df['Climate_category']
```

# ANALYSIS

```python
#1.Which countries have the highest and lowest populations?
print(df.loc[df['Population'].idxmax(), ['Country', 'Population']])
print()
print(df.loc[df['Population'].idxmin(), ['Country', 'Population']])
```

```python
# 2.What are the largest and smallest countries by land area?
df[['Country','Area (sq. mi.)']][(df['Area (sq. mi.)']==df['Area (sq. mi.)'].max())|(df['Area (sq. mi.)']==df['Area (sq. mi.)'].min())]
```

```python
#3.What are the top 10 countries by GDP and which country has the maximum GDP growth ?
df.nlargest(10, 'GDP ($ per capita)')[['Country', 'GDP ($ per capita)']]
```

```python
df[['Country','GDP ($ per capita)']][(df['GDP ($ per capita)']==df['GDP ($ per capita)'].max())]
```

```python
# 4.What is the average population density across all countries?
df['Pop. Density (per sq. mi.)'].mean()
```

```python
# 5.Which country is having the highest infant mortality rate?
df[['Country','Infant mortality (per 1000 births)']][df['Infant mortality (per 1000 births)']==df['Infant mortality (per 1000 births)'].max()]
```

```python
# 6.What is the average literacy rate by region?
df.groupby('Region')['Literacy (%)'].mean().sort_values(ascending=False)
```

```python
# 7.Which country has the lowest rates of mobile phone usage?
df[['Country','Phones (per 1000)']][df['Phones (per 1000)']==df['Phones (per 1000)'].min()]
```

```python
# 8.What is the birth and Death rate for each Continents, and find the continent with the highest birth and Death rate.
df.groupby('Continent')[['Birthrate', 'Deathrate']].mean()
```

```python
# 9.Literacy rate varies across different Continents?
df.groupby('Continent')['Literacy (%)'].mean()
```

```python
# 10.How many countries are included in the dataset?
df['Country'].nunique()
```

```python
# 11.Total Population by Region
df.groupby('Region')['Population'].sum().sort_values(ascending=False)
```

```python
# 12.Population Density Analysis
df['Population Density']= df['Population'] / df['Area (sq. mi.)']
df.nlargest(10, 'Population Density')[['Country', 'Population Density']]
```

```python
# 13. When considering the coastline, find the countries with the highest coastlines.
df[['Country','Coastline (coast/area ratio)']][df['Coastline (coast/area ratio)']==df['Coastline (coast/area ratio)'].max()]
```

```python
# 14. Find the countries with the most and least migrants.
df[['Country','Net migration']][(df['Net migration']==df['Net migration'].max())|(df['Net migration']==df['Net migration'].min())]
```

```python
# 15. Find the continent with the most and least net migrants.
continent_migration_sum=df.groupby('Continent')['Net migration'].sum()
print(continent_migration_sum)
print()
print('Continent with Most Migrants:',continent_migration_sum.idxmax(), continent_migration_sum.max())
print()
print('Continent with Least Migrants:',continent_migration_sum.idxmin(), continent_migration_sum.max())
```

```python
# 16. Average GDP per Capita by Continent
df.groupby('Continent')['GDP_per_Capita'].mean().sort_values(ascending=False)
```

```python
# 17. Highest GDP per Capita in Each Continent
df.loc[df.groupby('Continent')['GDP_per_Capita'].idxmax(),['Continent', 'Country', 'GDP_per_Capita']]
```

```python
# 18. Distribution of Density Categories
df['Density_category'].value_counts()
```
